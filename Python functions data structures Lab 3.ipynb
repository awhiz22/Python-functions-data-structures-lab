```python
print("running...")
records = [
    {"name":"Ava", "score": 95, "hours_studied": 10},
    {"name":"Ben", "score": None, "hours_studied": 8},
    {"name":"Cara", "score": 88, "hours_studied": None},
    {"name":"", "score": 72, "hours_studied": 6},
    {"name": "Eli", "score": 105, "hours_studied": 9},
    {"name": "Finn", "score": 65, "hours_studied": 7}, ]

def validate_records(records):
    required_keys = ["name", "score", "hours_studied"]
    errors = []

    for i, record in enumerate(records):
        for k in required_keys:
            if k not in record:
                errors.append(f"Row {i}: missing key '{k}'")

        for k, value in record.items():
            if value is None or value == "":
                errors.append(f"Row {i}: empty value for '{k}'")

        score = record.get("score")
        if score is None:
            errors.append(f"Row {i}: score is missing")
        elif not (0 <= score <= 100):
            errors.append(f"Row {i}: score out of range -> {score}")

        hours = record.get("hours_studied")
        if hours is None:
            errors.append(f"Row {i}: hours_studied is missing")
        elif hours < 0:
            errors.append(f"Row {i}: negative hours_studied")

        if not isinstance(record.get("name"), str):
            errors.append(f"Row {i}: name should be a string")

    print("errors found:", errors)

    if errors:
        for e in errors:
            print(e)
    else:
        print("All records valid!")

validate_records(records)
```

```python
def clean_names(records):
    cleaned = []
    for record in records:
        name = record.get("name")
        
        if name is None or name == "":
            print(f"Removed record with missing name: {record}")
            continue
        
        name = name.strip()
        
        name = name.title()
        
        name = ''.join(char for char in name if not char.isdigit())
        
        record["name"] = name
        cleaned.append(record)

    return cleaned
    


cleaned = clean_names(records)
for record in cleaned:
    print(record)
```

```python
def fill_missing_values(records):
    
    
    scores = [r["score"] for r in records if r.get("score") is not None]
    avg_score = sum(scores) / len(scores)
    
    
    hours = [r["hours_studied"] for r in records if r.get("hours_studied") is not None]
    avg_hours = sum(hours) / len(hours)
    
    print(f"Average score: {avg_score}")
    print(f"Average hours studied: {avg_hours}")
    
    
    for record in records:
        if record.get("score") is None:
            record["score"] = round(avg_score, 1)
            print(f"Replaced missing score in: {record['name']}")
            
        if record.get("hours_studied") is None:
            record["hours_studied"] = round(avg_hours, 1)
            print(f"Replaced missing hours in: {record['name']}")
    
    return records


filled = fill_missing_values(cleaned)
for record in filled:
    print(record)
```

```python
def fix_invalid_scores(records):
    for record in records:
        score = record.get("score")
        
        
        if score is None:
            continue
        
        
        if score > 100:
            print(f"Score {score} too high for {record['name']} — setting to 100")
            record["score"] = 100
        
        
        if score < 0:
            print(f"Score {score} negative for {record['name']} — setting to 0")
            record["score"] = 0
    
    return records

fixed = fix_invalid_scores(filled)
for record in fixed:
    print(record)
```
